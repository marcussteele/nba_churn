```python
def get_preds_and_f1(probs,y_test):
    thresh = get_threshold(probs,y_test)
    pred = (probs > thresh).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test,pred).ravel()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = 2 * (precision * recall)/(precision+recall)
    return pred,f1
```
